# **TẢI CÁC THƯ VIỆN VÀ ĐỌC ẢNH**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Download image from internet
!wget https://image.tienphong.vn/w1000/Uploaded/2024/lzwlzsg-eafzljmgfy/2023_01_02/ronaldo-thumb-0101002922-7307.jpg
```

```python
# Read image
img_original = cv2.imread("/content/ronaldo-thumb-0101002922-7307.jpg")
# Convert Color Space into Red, Green, Blue
img_original = cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB)
img_original.shape
```

```python
# Plot image
def plot_image(img_ori, red, green, blue, title1, title2, title3, title4):
  plt.figure(figsize = (10,10))
  plt.subplot(141)
  plt.title(title1)
  plt.imshow(img_ori)
  plt.axis("off")
  plt.subplot(142)
  plt.title(title2)
  plt.imshow(red, cmap ="gray")
  plt.axis("off")
  plt.subplot(143)
  plt.title(title3)
  plt.imshow(green, cmap ="gray")
  plt.axis("off")
  plt.subplot(144)
  plt.title(title4)
  plt.imshow(blue, cmap ="gray")
  plt.axis("off")
  plt.show()
```

# **YÊU CẦU 1: TÁCH ẢNH XÁM TỪ ẢNH MÀU**

```python
# Split color channel
r,g,b = cv2.split(img_original)
print(f"Red shape: {r.shape}")
print(f"Green shape: {g.shape}")
print(f"Blue shape: {b.shape}")
```

```python
# Plot color image and R,G,B channel in gray scale
plot_image(img_original, r, g, b, "Color", "Red Channel", "Green Channel", "Blue Channel")
```

# **YÊU CẦU 2: KẾT HỢP CÁC CHANNEL MÀU ĐỂ TẠO THÀNH ẢNH MÀU**

```python
# Merge red, green, blue channel into color image
img_color = cv2.merge([r,g,b])
print(f"Image Color: {img_color.shape}")
```

```python
plt.imshow(img_color)
plt.title("Color Image")
plt.axis("off")
plt.show
```
