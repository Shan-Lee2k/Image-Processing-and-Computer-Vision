# **TOPIC 1: LOW-PASS FILTER**

**ĐỌC ẢNH VÀ ÁP DỤNG BỘ LỌC GAUSSIAN**

```python
lenna = cv2.imread("/content/saltandpepperlena-300x300.jpg",cv2.IMREAD_GRAYSCALE)
print(lenna.shape)
blur_gauss = cv2.GaussianBlur(lenna,(11,11),3)
plt.subplot(121),plt.imshow(lenna, cmap ="gray"),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur_gauss, cmap ="gray"),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
boat = cv2.imread("/content/boat1.png",cv2.IMREAD_GRAYSCALE)
print(lenna.shape)
blur_gauss_boat = cv2.GaussianBlur(boat,(9,9),3,3)
plt.subplot(121),plt.imshow(boat, cmap ="gray"),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur_gauss_boat, cmap ="gray"),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

# **TOPIC 2: HIGH-PASS FILTER**

```python
def high_pass_filter(image):

  # Làm mờ ảnh bằng hàm Gauss
  blurred_image = cv2.GaussianBlur(image, (9, 9), 3)

  # Lấy ảnh gốc trừ đi ảnh đã được làm mờ
  high_pass_image = image - blurred_image

  return high_pass_image
```

```python
kernel = np.array([[-1, -1, -1],
                    [-1,  8, -1],
                    [-1, -1, -1]])
high_pass_boat = high_pass_filter(boat)
plt.subplot(121),plt.imshow(boat, cmap ="gray"),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(high_pass_boat, cmap ="gray"),plt.title('High-Pass Filter')
plt.xticks([]), plt.yticks([])
plt.show()
```
